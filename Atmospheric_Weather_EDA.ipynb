## **DATA CLEANING AND PREPROCESSING**

```python
missing_values = data.isnull().sum()
print(missing_values[missing_values > 0]) # Show only columns that have at least one missing value
```

## Handles missing values

```python
data = data.drop_duplicates()
print("Duplicate data removed!") 
```

## Remove duplicate rows

```python
numeric_columns = data.select_dtypes(include=['int64', 'float64']).columns
numeric_cols_with_na = [c for c in numeric_columns if data[c].isnull().any()]
data[numeric_cols_with_na] = data[numeric_cols_with_na].fillna(data[numeric_cols_with_na].median())
# Only touches columns that actually had missing numeric values (e.g. temperature, humidity),
# instead of blindly rewriting every numeric column in the dataset.
print(f"Filled missing values in: {numeric_cols_with_na or 'no numeric columns needed it'}")
```

## Handles missing texts

```python
categorical_cols = data.select_dtypes(include=['object']).columns
cat_cols_with_na = [c for c in categorical_cols if data[c].isnull().any()]

# Be careful with identifying columns like 'country' or 'location_name':
# silently filling these with the most frequent value would distort the
# geographic analysis, so we flag them instead of auto-filling.
key_id_cols = {'country', 'location_name'}
risky = [c for c in cat_cols_with_na if c in key_id_cols]
if risky:
    print(f"Warning: {risky} have missing values — review these manually before filling.")

safe_cols = [c for c in cat_cols_with_na if c not in key_id_cols]
for col in safe_cols:
    data[col] = data[col].fillna(data[col].mode()[0])

print(f"Missing values safely handled for: {safe_cols or 'no categorical columns needed it'}")
```

## Fix the date column format

```python
data['last_updated'] = pd.to_datetime(data['last_updated'])
print("'last_updated' column converted to proper Date format.") # This converts the text dates into actual datetime timestamps so Python can plot timelines
```

```python
print("\n--- Data Cleaning Complete! ---")
print(f"Your ready-to-use dataset shape is: {data.shape}")
```

```python
remaining_nulls = data.isnull().sum().sum()
print(f"Remaining missing values in the whole dataset: {remaining_nulls}")
assert remaining_nulls == 0, "There are still missing values — check the columns above."
```

## **Correlation Heatmap**

```python
core_features = [
    'temperature_celsius', 'feels_like_celsius', 'wind_kph', 'pressure_mb', 
    'precip_mm', 'humidity', 'cloud', 'visibility_km', 'uv_index', 
    'air_quality_Carbon_Monoxide', 'air_quality_Ozone', 
    'air_quality_Nitrogen_dioxide', 'air_quality_Sulphur_dioxide'
]
```

```python
# Calculate the correlation matrix
corr_matrix = data[core_features].corr()
# Create the heatmap plot
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=0.5)
plt.title('Correlation Heatmap of Meteorological and Air Quality Parameters', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## **Outlier Detection (Boxplots)**

```python
# List the 5 key columns we want to check for anomalies
outlier_params = ['temperature_celsius', 'humidity', 'wind_kph', 'uv_index', 'pressure_mb']

fig, axes = plt.subplots(2, 3, figsize=(16, 10))
# Flatten the axes array from a grid to a single line so it's easier to loop through
axes = axes.flatten()

# Loop through our parameters and draw a boxplot for each
for i, param in enumerate(outlier_params):
    if param in data.columns:
        sns.boxplot(y=data[param], ax=axes[i], color='lightgreen')
        axes[i].set_title(f'Outliers in {param}', fontsize=12, fontweight='bold')
    axes[i].set_ylabel('')

# Since we only have 5 parameters but a 2x3 grid has 6 spots, 
# let's turn off the empty 6th spot so it looks perfectly clean.
axes[-1].axis('off')

plt.tight_layout()
plt.show()
```

## **Geospatial Visualisation — Folium Interactive Map**

```python
import folium
```

```python
print("Filtering data for the latest unique locations...") 
```

```python
latest_data = data.sort_values('last_updated').groupby('location_name').last().reset_index()

print("Initializing the world map...")
```

```python
mymap = folium.Map(location=[20, 0], zoom_start=2, tiles='OpenStreetMap')
```

```python
# Create a simple helper function to choose colors based on how hot it is
def get_temperature_color(temp):
    if temp < 10:
        return 'blue'       # Cold
    elif temp < 25:
        return 'green'      # Mild
    elif temp < 35:
        return 'orange'     # Hot
    else:
        return 'red'        # Very Hot
print("Adding interactive markers to the map")
```

```python
for idx, row in latest_data.iterrows():
    if pd.notna(row['latitude']) and pd.notna(row['longitude']):
        # Set up what the popup text will say when you click a circle
        popup_info = f"<b>Country:</b> {row['country']}<br><b>City:</b> {row['location_name']}<br><b>Temp:</b> {row['temperature_celsius']}°C"
        
        folium.CircleMarker(
            location=[row['latitude'], row['longitude']],
            radius=5,
            popup=folium.Popup(popup_info, max_width=300),
            color=get_temperature_color(row['temperature_celsius']),
            fill=True,
            fill_color=get_temperature_color(row['temperature_celsius']),
            fill_opacity=0.7
        ).add_to(mymap)
```

```python
# Save the map as an interactive HTML file you can open in your browser!
mymap.save('global_weather_map.html')
print("Map successfully built and saved as 'global_weather_map.html'!")
mymap
```

```python
import plotly.express as px
```

```python
print("Calculating average temperature per country (based on each location's most recent reading)...")
# NOTE: latest_data holds only the single most recent reading per location, so this
# is the average of the latest snapshot across locations in a country -- not a
# true multi-day average temperature.
country_data = latest_data.groupby('country')['temperature_celsius'].mean().reset_index()
```

```python
print("Building the Plotly Choropleth map...")
# Create the interactive choropleth map
fig = px.choropleth(
    country_data,
    locations="country",
    locationmode="country names",
    color="temperature_celsius",
    hover_name="country",
    color_continuous_scale=px.colors.sequential.Plasma,
    title="Global Temperature by Country (Latest Reading, Averaged Across Locations)",
    labels={'temperature_celsius': 'Avg Temp (°C)'}
)
```

```python
fig.update_layout(
    geo=dict(showframe=False, showcoastlines=True, projection_type='equirectangular'),
    margin={"r":0,"t":40,"l":0,"b":0}
)

# Save it as an interactive webpage file too
fig.write_html("global_weather_choropleth.html")
print("Choropleth successfully built and saved as 'global_weather_choropleth.html'!")
fig.show()
```
